# Calculator

```python
a=eval(input("Enter 1st value:"))
o=input("Enter Operator(+,-,*,/):")
b=eval(input("Enter 2nd value:"))
if o=="+":
    print(a,"+",b,"=",a+b)
elif o=="-":
    print(a,"-",b,"=",a-b)
elif o=="*":
    print(a,"*",b,"=",a*b)
elif o=="/":
    print(a,"/",b,"=",a/b)
else:
    print("Enter appropriate operator :)! ")
```

# Q7


```python
Eng = int(input("Enter English Marks:"))
Phy = int(input("Enter Physics Marks:"))
Math = int(input("Enter Maths Marks:"))
Urdu = int(input("Enter Urdu Marks:"))
Comp = int(input("Enter Comp Marks:"))
obtain_marks = Eng + Phy + Math + Urdu + Comp
print("Your Obtain Marks is : ",obtain_marks)
per = obtain_marks / 500 * 100
print("Your Percentage is : ", per)
if per >=80 and per <=100:
    print("A-One Grade")
elif per >= 70 and per < 80:
    print("A Grade")
elif per >= 60 and per < 70:
    print("B Grade")
elif per >= 50 and per < 60:
    print("C Grade")
elif per >= 40 and per < 50:
    print("D Grade")
elif per >= 33 and per < 40:
    print("E Grade")
else:
    print("Fail")
```
