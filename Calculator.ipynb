```python
a=eval(input("Enter 1st value:"))
o=input("Enter Operator(+,-,*,/):")
b=eval(input("Enter 2nd value:"))
if o=="+":
    print(a,"+",b,"=",a+b)
elif o=="-":
    print(a,"-",b,"=",a-b)
elif o=="*":
    print(a,"*",b,"=",a*b)
elif o=="/":
    print(a,"/",b,"=",a/b)
else:
    print("Enter appropriate operator :)! ")
```
